## Public Holiday API

```python
import requests
import json     #importing necessary packages
import csv
import pandas as pd

# Country Codes
countryCode = "US"
url = f"https://date.nager.at/api/v3/CountryInfo/{countryCode}" # accessing specific API link and calling the country code
country_data = requests.get(url) # getting the data from the url
json_file = country_data.json() # collects the basic details of a specified country

json_file
```

```python
# Available Countries

url = f"https://date.nager.at/api/v3/AvailableCountries"
available_data = requests.get(url)
json_file = available_data.json() 
json_file # accesses the complete list of available countries and their corresponding code
```

```python
# Long Weekends

countryCode = "US"
Year = "2022"
url = f"https://date.nager.at/api/v3/LongWeekend/{Year}/{countryCode}" 
weekend_data = requests.get(url) 
json_file = weekend_data.json() 
json_file
# identifies the dates of long weekends during a specific year in a specific country
```

```python
# Holidays

countryCode = "US"
Year = "2020"
url = f"https://date.nager.at/api/v3/PublicHolidays/{Year}/{countryCode}" 
holiday_data = requests.get(url) 
json_file = holiday_data.json() 
json_file
# uses specified year and code to identify all of the details of holidays in a given country
```

```python
# Next Public Holiday

countryCode = "US"
url = f"https://date.nager.at/api/v3/NextPublicHolidays/{countryCode}" 
next_data = requests.get(url) 
json_file = next_data.json() 
json_file
# based on location and date data, accesses a complete list of upcoming public holidays
```
